# Bronze Layer

```python
from pyspark.sql.types import StructType , StructField , StringType
customer_schema=StructType([
    StructField("customer_id",StringType(),True),
    StructField("name",StringType(),True),
    StructField("email",StringType(),True),
    StructField("phone",StringType(),True),
    StructField("location",StringType(),True),
    StructField("signup_date",StringType(),True),
    StructField("is_premium",StringType(),True),
    StructField("preferred_cuisine",StringType(),True),
    StructField("total_orders",StringType(),True),
    StructField("average_rating",StringType(),True)
])

orders_schema=StructType([
    StructField("order_id",StringType(),True),
    StructField("customer_id",StringType(),True),
    StructField("restaurant_id",StringType(),True),
    StructField("order_date",StringType(),True),
    StructField("delivery_time",StringType(),True),
    StructField("status",StringType(),True),
    StructField("total_amount",StringType(),True),
    StructField("payment_mode",StringType(),True),
    StructField("discount_applied",StringType(),True),
    StructField("feedback_rating",StringType(),True)
])
  
restaurants_schema=StructType([
    StructField("restaurant_id",StringType(),True),
    StructField("name",StringType(),True),
    StructField("cuisine_type",StringType(),True),
    StructField("location",StringType(),True),
    StructField("owner_name",StringType(),True),
    StructField("average_delivery_time",StringType(),True),
    StructField("contact_number",StringType(),True),
    StructField("rating",StringType(),True),
    StructField("total_orders",StringType(),True),
    StructField("is_active",StringType(),True),
    StructField("opening_date",StringType(),True)
])

deliveries_schema=StructType([
    StructField("delivery_id",StringType(),True),
    StructField("order_id",StringType(),True),
    StructField("delivery_person_id",StringType(),True),
    StructField("distance",StringType(),True),
    StructField("delivery_time",StringType(),True),
    StructField("estimated_time",StringType(),True),
    StructField("delivery_fee",StringType(),True),
    StructField("vehicle_type",StringType(),True)
])

delivery_persons_schema=StructType([
    StructField("delivery_person_id",StringType(),True),
    StructField("name",StringType(),True),
    StructField("contact_number",StringType(),True),
    StructField("vehicle_type",StringType(),True),
    StructField("total_deliveries",StringType(),True),
    StructField("average_rating",StringType(),True),
    StructField("location",StringType(),True)
])
```

### Reading Raw as CSV

```python
orders_bronze_df=spark.read.format("csv").option("header","true").load("/Volumes/zomato_project/raw_data/raw_files/orders.csv")
customers_bronze_df=spark.read.format("csv").option("header","true").load("/Volumes/zomato_project/raw_data/raw_files/customers.csv")
restaurants_bronze_df=spark.read.format("csv").option("header","true").load("/Volumes/zomato_project/raw_data/raw_files/restaurants.csv")
deliveries_bronze_df=spark.read.format("csv").option("header","true").load("/Volumes/zomato_project/raw_data/raw_files/deliveries.csv")
delivery_persons_bronze_df=spark.read.format("csv").option("header","true").load("/Volumes/zomato_project/raw_data/raw_files/delivery_persons.csv")
```

```python
orders_bronze_df.display()
customers_bronze_df.display()
restaurants_bronze_df.display()
deliveries_bronze_df.display()
delivery_persons_bronze_df.display()
```

### Add Ingestion metadata

```python
from pyspark.sql.functions import current_timestamp
orders_bronze_df=orders_bronze_df.withColumn('ingest_timestamp',current_timestamp())
customers_bronze_df=customers_bronze_df.withColumn('ingest_timestamp',current_timestamp())
restaurants_bronze_df=restaurants_bronze_df.withColumn('ingest_timestamp',current_timestamp())
deliveries_bronze_df=deliveries_bronze_df.withColumn('ingest_timestamp',current_timestamp())
delivery_persons_bronze_df=delivery_persons_bronze_df.withColumn('ingest_timestamp',current_timestamp())
```

### Save As Delta Table in Bronze

```python
orders_bronze_df.write.format("delta").mode("overwrite").saveAsTable('zomato_project.bronze.bronze_orders')
customers_bronze_df.write.format("delta").mode("overwrite").saveAsTable('zomato_project.bronze.bronze_customers')
restaurants_bronze_df.write.format("delta").mode("overwrite").saveAsTable('zomato_project.bronze.bronze_restaurants')
deliveries_bronze_df.write.format("delta").mode("overwrite").saveAsTable('zomato_project.bronze.bronze_deliveries')
delivery_persons_bronze_df.write.format("delta").mode("overwrite").saveAsTable('zomato_project.bronze.bronze_delivery_persons')
```
